```python
z=zeros((100,100))
n,m=100,100
for i in range(n):
    for j in range(m):
        a=-3+i*6/99
        b=-3+j*6/99
        z[i][j] = exp(-a**2-b**2)
```

```python
#for i in range(100):
#    z[i,0] = z[i,-1]=0
#for i in range(100):
#    z[0,j] = z[-1,j]=0
z[:,-1]=z[:,0]=0
z[0,:]=z[-1,:]=0

imshow(z,cmap='cividis')
show()
plot(z[49,:])
show()
```

## Exercice 4 PIL et Numpy

```python
from PIL import Image
barb=array(Image.open('img/barbara.png'))
n,m=barb.shape
imshow(barb)
show()
```

```python
plot(barb[n//2,:])
show()
```

```python
I1=barb[:n//2,:m//2]
I2=barb[:n//2,m//2:]
I3=barb[n//2:,:m//2]
I4=barb[n//2:,m//2:]

subplot(2,2,1)
imshow(I1)
subplot(2,2,2)
imshow(I2)
subplot(2,2,3)
imshow(I3)
subplot(2,2,4)
imshow(I4)
show()
```

```python
I=vstack((hstack((I1,I2)),hstack((I3,I4))))
imshow(I)
```

```python
J=I[::2,::2]
imshow(J)
I.shape,J.shape
```
